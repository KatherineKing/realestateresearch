```python
# load required packages
import pandas as pd
```

```python
df = pd.read_csv("cobb-county-ga-addresses.csv")
output_df = df[["CompleteStreetName", "ZipCode", "City", "State"]]
output_df = output_df.rename(columns={
    "City": "CITY",
    "State": "STATE",
    "CompleteStreetName": "ADDRESS",
    "ZipCode": "ZIP OR POSTAL CODE"
}).dropna()
output_df["ADDRESS"] = output_df.apply(lambda x: x["ADDRESS"].upper(), axis=1)
output_df["CITY"] = output_df.apply(lambda x: x["CITY"].upper(), axis=1)
output_df["STATE"] = output_df.apply(lambda x: x["STATE"].upper(), axis=1)
output_df.to_csv("cobb-county-ga-addresses-PARSED.csv", index=False)
output_df.head(10)
```
